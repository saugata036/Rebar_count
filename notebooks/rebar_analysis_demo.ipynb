```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

IMAGE_PATH = PROJECT_ROOT / "input_images"/ "clean_front_face_output_4.png"
OUTPUT_DIR = PROJECT_ROOT / "outputs" / "test"

print(f"Project root: {PROJECT_ROOT}")
print(f"Input image:  {IMAGE_PATH}")
```

```python
import json
from rebar_xray.pipeline import RebarXRayPipeline
from rebar_xray.config import PipelineConfig

result = RebarXRayPipeline(PipelineConfig()).run(
    IMAGE_PATH,
    output_dir=OUTPUT_DIR,
).payload

print(json.dumps(result, indent=2))
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

from rebar_xray.utils import load_bgr


def read_rgb(path) -> np.ndarray:
    return cv2.cvtColor(load_bgr(path), cv2.COLOR_BGR2RGB)


original = read_rgb(IMAGE_PATH)
counted = read_rgb(result["output_image"])
counts = result["bar_counts"]

print(f"Vertical bars:   {counts['vertical_rods']}")
print(f"Horizontal bars: {counts['horizontal_rods']}")

fig, axes = plt.subplots(1, 2, figsize=(16, 8))
axes[0].imshow(original)
axes[0].set_title("Front-face input")
axes[0].axis("off")
axes[1].imshow(counted)
axes[1].set_title("Counted bars")
axes[1].axis("off")
plt.tight_layout()
plt.show()
```
